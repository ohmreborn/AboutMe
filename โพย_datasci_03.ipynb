```python
import pandas as pd
data = {
 "calories": [420, 380, 390],
 "duration": [50, 40, 45]
}
df = pd.DataFrame(data)
""" output df
   calories  duration
0       420        50
1       380        40
2       390        45
"""
#print ตารางออกมาแค่ตําแหน่งที่ 0
"""
output df.loc[0]
calories    420
duration     50
"""
```

```python
import pandas as pd
data = {
 "calories": [420, 380, 390],
 "duration": [50, 40, 45]
}

df = pd.DataFrame(data, index = ["day1", "day2", "day3"])
"""   output df
      calories  duration
day1       420        50
day2       380        40
day3       390        45
"""
```
