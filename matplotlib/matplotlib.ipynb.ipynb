```python
import matplotlib.pyplot as plt
```

```python
x = [0,5,7]
y = [0, 200,10]

plt.title("hello")
plt.plot(x,y)
plt.grid()

plt.show()
```

```python
x = ["america","thai","japan"]
y = [1,3,5]
plt.bar(x,y, color= "red")
plt.show()
```

```python
x = ["a","b","c","d"]
y = [1,3,5,2]

plt.barh(x,y, color = ["#9360b9","red", "yellow","purple"])
plt.show()
```

```python
y = [20,30,25]
lb = ["gold","silver","bronze"]
plt.pie(y,labels=lb,autopct="%1.2f%%")
plt.show()
```

```python
y = [20,30,25,25]
lb = ["a","b","c","d"]
plt.figure(figsize=(10,10))
plt.pie(y, labels=lb, autopct='%.1f%%')
plt.show()
```
